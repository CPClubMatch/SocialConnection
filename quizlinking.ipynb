```python
import pandas as pd
import numpy as np
```

```python
clubscored_df = pd.read_csv("ClubsScored.csv")
clubs_df = pd.read_csv("NicosScrapedData.csv")
print(clubscored_df.head(10))
```

```python
threshold = 0.7
highestN = 3

#new_df = clubscored_df.drop(columns=["Club Name"])
new_df = clubscored_df
top3 = new_df.iloc[:, :-1].apply(lambda row: row.nlargest(3).values, axis=1)

print(top3)
```

```python
pd.set_option('display.max_columns', None)
top3 = []
top6 = []
club_names = new_df["Club Name"]
for index, row in new_df.iloc[:, :-1].iterrows():
    top3.append(row.nlargest(3).index.tolist())
    top6.append(row.nlargest(6).index.tolist())

final_df = new_df.iloc[:, :-1].mask(new_df.iloc[:, :-1] < threshold, 0)
#final_df = df.mask((df >= threshold) & (df in top6), 1)

for i in range(len(final_df)):
    for col in top6[i]:
        if col in final_df.columns and col in top3[i]:
            final_df.loc[i, col] = 1
        elif col in final_df.columns and final_df.loc[i, col] > threshold:
            final_df.loc[i, col] = 1
        else:
            final_df.loc[i, col] = 0
final_df["Club Name"] = club_names
final_df
```

```python

pd.set_option('display.max_columns', None)
top3 = []
club_names = new_df["Club Name"]
for index, row in new_df.iloc[:, :-1].iterrows():
    top3.append(row.nlargest(3).index.tolist())

df = new_df.iloc[:, :-1].mask(new_df.iloc[:, :-1] < threshold, 0)
final_df2 = df.mask(df >= threshold, 1)

for i in range(len(final_df2)):
    for col in top3[i]:
        if col in final_df2.columns:
            final_df.loc[i, col] = 1
final_df2["Club Name"] = club_names
final_df2
```

```python
tag_list = final_df.iloc[:, :-1].columns
tags_df = pd.merge(final_df, clubs_df.drop(columns=['tablescraper-selected-row href']), on="Club Name", how="left")
tags_df
```

```python
export_df = tags_df[["Club Name", "Description Excerpt"]]
export_df["Tags"] = tags_df.apply(
    lambda row: [col for col in tag_list if row[col] !=0],
    axis = 1)
export_df
```

```python
export_df.to_csv("quiz_format.csv", index=False)
```
